```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
```

```python
link ="https://www.screener.in/company/RELIANCE/"
```

```python
response = requests.get(link)

if(response.status_code == 200):
  print("Success")
elif(response.status_code == 404):
  print("404")
else:
  print("Status Code:",response.status_code)
```

with open('sample_screener.html','w') as f1:
  f1.write(response.text)

```python
text =""
with open('sample_screener.html','r') as f1:
  text = f1.read()
```

```python
soup = bs(text,'html.parser')

parent = soup.find('section',id='documents')

child = parent.find_all('a')

href =[]
text = []

for i,j in enumerate(child):
  href.append(j.get('href'))
  text.append(j.text.strip())

```

```python
df = pd.DataFrame({"text":text,'href':href})
df
```
