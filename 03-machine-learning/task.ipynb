```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from sklearn.preprocessing import  PolynomialFeatures
```

```python
x = np.array([5,15,25,35,45,55]).reshape(-1, 1)
y = np.array([5,20,14,32,22,38])


model = LinearRegression()
model.fit(x,y)
pre = model.predict(x)

acc = r2_score(y,pre)
mse = mean_squared_error(y,pre)
print("Accuracy:",acc*100)
print("Mean Squared Error:",mse)
print("Model Intercept:",model.intercept_)
print("Model Coeff:",model.coef_)
x= np.arange(5).reshape(-1,1)
pre = model.predict(x)
print(pre)
```

```python
df = pd.read_csv("/home/shlok/College/SemIV/FCSP2/03-machine-learning/DATA/DATASET/Datasets/car data.csv")

df['Age_car'] = 2023 - df['Year']

print(df.info())

df = pd.get_dummies(df,columns=['Fuel_Type','Transmission','Seller_Type'],drop_first=True)

features = ['Age_car','Kms_Driven']
features.extend([col for col in df.columns if col.startswith('Fuel_Type_') or col.startswith('Transmission_')or col.startswith('Seller_Type_')])

x = df[features]
y = df['Selling_Price']


x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)


model = LinearRegression()
model.fit(x_train, y_train)
pre = model.predict(x_test)

acc = r2_score(y_test, pre)
mse = mean_squared_error(y_test, pre)
print("Accuracy (R2 Score):", acc * 100)
print("Mean Squared Error:", mse)

entry = df[df['Car_Name']== 'Hero Super Splendor'][features]
print("entry: ",model.predict(entry))

entry = df.loc[[5]][features]
print("entry: ",model.predict(entry))
```

```python
df = pd.read_csv('/home/shlok/College/SemIV/FCSP2/03-machine-learning/DATA/DATASET/Datasets/winequalityN.csv')


df.fillna(df.mean(numeric_only=True), inplace=True)


df = pd.get_dummies(df, columns=['type'], drop_first=True)

x = df.drop(columns=['quality'])
y = df['quality']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(x_train, y_train)

print(f"Co-efficients: {model.coef_}")
print(f"Intercept: {model.intercept_}")

pre = model.predict(x_test)
print(f"Mean Squared Error: {mean_squared_error(y_test, pre)}")
print(f"R2 Score: {r2_score(y_test, pre) * 100:.2f}%")

new_data = pd.DataFrame({
    'fixed acidity': [8],
    'volatile acidity': [0.4],
    'citric acid': [0.40],
    'residual sugar': [15],
    'chlorides': [0.048],
    'free sulfur dioxide': [40],
    'total sulfur dioxide': [150],
    'density': [0.99],
    'pH': [3],
    'sulphates': [0.45],
    'alcohol': [10.5],
    'type_white': [0] 
})

prediction = model.predict(new_data)
print(f"\nPredicted Quality for provided Red Wine data: {prediction[0]:.2f}")
```
