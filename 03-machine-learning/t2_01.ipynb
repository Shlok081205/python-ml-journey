```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from sklearn.preprocessing import  PolynomialFeatures
```

# Linear Regression

```python
df = pd.read_csv("/home/shlok/College/SemIV/FCSP2/03-machine-learning/DATA/DATASET/Datasets/heights_weights.csv")
print("Null Values:",df.isnull().sum().sum())
print("Columns Name:",df.columns)

x = df[['Height']]
y= df['Weight']

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)

print("X Train:",x_train.shape," Y Train:",y_train.shape)
print("X Test:",x_test.shape," Y Test:",y_test.shape)


model = LinearRegression()
model.fit(x_train,y_train)
pre = model.predict(x_test)

acc = r2_score(y_test,pre)
mse = mean_squared_error(y_test,pre)
print("Accuracy:",acc*100)
print("Mean Squared Error:",mse)
print("Model Intercept:",model.intercept_)
print("Model Coeff:",model.coef_)


plt.scatter(x_train, y_train, color='green', label='Actual', alpha=0.7)
plt.scatter(x_test, y_test, color='blue', label='Actual', alpha=0.7)
plt.plot(x_test,pre, color='red', linewidth=2, label='Predicted')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Simple Linear Regression')
plt.legend()
plt.show()
```

# Polynomial Regression

```python
df = pd.read_csv("/home/shlok/College/SemIV/FCSP2/03-machine-learning/DATA/DATASET/Datasets/Fish.csv")

print("Null Values:", df.isnull().sum().sum())
print("Columns Name:", df.columns)


features = ['Length1', 'Length2', 'Length3', 'Height', 'Width']
x = df[features]
y = df['Weight']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)


poly = PolynomialFeatures(degree=3, include_bias=False) # Degree 3
x_train_poly = poly.fit_transform(x_train) # 'fit_transform'
x_test_poly = poly.transform(x_test) # Use .transform() on test data

model = LinearRegression()
model.fit(x_train_poly, y_train)

pre = model.predict(x_test_poly)

acc = r2_score(y_test, pre)
mse = mean_squared_error(y_test, pre)

print("Accuracy (R2 Score):", acc * 100)
print("Mean Squared Error:", mse)
print("Model Intercept:",model.intercept_)
print("Model Coeff:",model.coef_)
```

```python
df = pd.read_csv("/home/shlok/College/SemIV/FCSP2/03-machine-learning/DATA/DATASET/Datasets/Fish.csv")


features = ['Length1', 'Length2', 'Length3', 'Height', 'Width']
x = df[features]
y = df['Weight']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
arr =[]

best_acc = -float('inf')
patience = 5
wait = 0
count = 1 

while True:
    poly = PolynomialFeatures(degree=count, include_bias=False) 
    x_train_poly = poly.fit_transform(x_train) 
    x_test_poly = poly.transform(x_test) 

    model = LinearRegression()
    model.fit(x_train_poly, y_train)

    pre = model.predict(x_test_poly)
    acc = r2_score(y_test, pre)
    mse = mean_squared_error(y_test, pre)

    arr.append({'acc': acc, 'mse': mse, "degree": count})

    
    if acc > best_acc:
        best_acc = acc
        wait = 0
    else:
        wait += 1

    if wait >= patience:
        print(f"Stopping. Max accuracy {best_acc:.4f} found at degree {count - patience}")
        break
    
    count += 1


```

```python
df = pd.read_csv("/home/shlok/College/SemIV/FCSP2/03-machine-learning/DATA/DATASET/Datasets/Fish.csv")


features = ['Length1', 'Length2', 'Length3', 'Height', 'Width']
x = df[features]
y = df['Weight']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

degrees = list(range(1, 21))
accuracies = []
results = []

for d in degrees:
    poly = PolynomialFeatures(degree=d, include_bias=False) 
    x_train_poly = poly.fit_transform(x_train) 
    x_test_poly = poly.transform(x_test) 

    model = LinearRegression()
    model.fit(x_train_poly, y_train)

    pre = model.predict(x_test_poly)
    acc = r2_score(y_test, pre)
    mse = mean_squared_error(y_test, pre)
    
    # Store for printing/plotting
    # We clip accuracy at -1 for the graph to prevent extreme outliers from ruining the scale
    accuracies.append(max(acc, -1)) 
    results.append({'degree': d, 'acc': acc, 'mse': mse})

# Find the best degree
best_result = max(results, key=lambda x: x['acc'])
print(f"Maximum Accuracy of {best_result['acc']:.4f} found at Degree {best_result['degree']}")

# Plotting the Bar Graph
plt.figure(figsize=(12, 6))
plt.bar(degrees, accuracies, color='skyblue', edgecolor='navy')
plt.xlabel('Polynomial Degree')
plt.ylabel('R2 Score (Accuracy)')
plt.title('Model Accuracy vs. Polynomial Degree (1 to 20)')
plt.xticks(degrees)
plt.ylim(-1.1, 1.1)  # Keeps the focus on the positive accuracy range
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

  
```
