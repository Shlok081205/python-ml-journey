```
### Your second question — where do min and max come from?

You are not finding a column called min or max. You are **applying the min and max functions** to the existing `Close Price` values within each group.

Think of it this way. After `groupby("Year")`, Pandas internally creates two buckets:
```
Year 2025 bucket:
[3624.95, 3568.00, 3529.95, 3483.95 ... all 2025 closing prices]

Year 2026 bucket:
[2636.40, 2647.70, 2629.50, 2574.50 ... all 2026 closing prices]
```

Then `agg()` runs your chosen functions on each bucket:
```
2025 bucket → mean() → 3256.37
2025 bucket → max()  → 3657.70
2025 bucket → min()  → 2889.00

2026 bucket → mean() → 3008.24
2026 bucket → max()  → 3294.45
2026 bucket → min()  → 2574.50


## Concept 9: merge() and concat()

### concat() — Stack DataFrames:
```python
# Vertical (more rows)
pd.concat([df1, df2], axis=0, ignore_index=True)

# Horizontal (more columns)
pd.concat([df1, df2], axis=1)
```
- Always use ignore_index=True when stacking vertically
  to avoid duplicate index values

### merge() — Join on a Common Column:
```python
pd.merge(df1, df2, on="common_column", how="join_type")
```

### Join Types:
| how    | Result                                          |
|--------|-------------------------------------------------|
| inner  | Only rows present in BOTH DataFrames            |
| left   | All rows from left, NaN where right has no match|
| right  | All rows from right, NaN where left has no match|
| outer  | All rows from both, NaN where either has no match|

### Key Points:
- merge() is equivalent to SQL JOIN
- inner join is the safest — no NaN values introduced
- Always run isnull().sum() after a merge to check for NaN
- outer join can introduce many NaN values unexpectedly
- on= parameter must be a column that exists in both DataFrames

````python
#Task-9:
"""
Stack both DataFrames vertically using concat() with ignore_index=True
Merge both DataFrames using an inner join on Product
Merge both DataFrames using a left join on Product and check for NaN values
"""

q1_sales = pd.DataFrame({
    "Product": ["Laptop", "Phone", "Tablet"],
    "Q1_Revenue": [150000, 80000, 45000]
})

q2_sales = pd.DataFrame({
    "Product": ["Phone", "Tablet", "Watch"],
    "Q2_Revenue": [95000, 50000, 30000]
})


print(pd.concat([q1_sales,q2_sales],axis=0,ignore_index=True),"\n")

print(pd.merge(q1_sales,q2_sales,on="Product",how="inner"),"\n")

print(pd.merge(q1_sales,q2_sales,on="Product",how="left").isnull().sum())
````

## Concept 10: Two-Way Cross Tabulation

### Basic syntax:
```python
pd.crosstab(df["Row_Variable"], df["Column_Variable"])
```

### With totals:
```python
pd.crosstab(df["Row"], df["Col"], margins=True)
```

### Normalized (proportions):
```python
# Whole table proportions
pd.crosstab(df["Row"], df["Col"], normalize=True)

# Row-wise proportions (each row sums to 1)
pd.crosstab(df["Row"], df["Col"], normalize="index")

# Column-wise proportions (each column sums to 1)
pd.crosstab(df["Row"], df["Col"], normalize="columns")
```

### Key Points:
- crosstab is used for two CATEGORICAL variables
- margins=True adds row and column totals
- normalize="index" shows pass/fail rate per category — 
  useful for comparing groups of different sizes
- Foundation of chi-square statistical tests

````python
#Task-10:
data = pd.DataFrame({
    "Gender":     ["Male","Female","Male","Female","Male",
                   "Female","Male","Female","Male","Female"],
    "Department": ["CE","IT","CE","CE","IT",
                   "IT","CE","IT","CE","CE"],
    "Result":     ["Pass","Pass","Fail","Pass","Pass",
                   "Fail","Pass","Pass","Fail","Pass"]
})
"""
Create a cross tabulation between Gender and Result
Add margins to see totals
Create a cross tabulation between Department and Result normalized by row — so you can see pass/fail rate per department as a proportion
"""

print(pd.crosstab(data["Gender"],data["Result"],margins=True),"\n")

print(pd.crosstab(data["Department"],data["Result"],normalize="index"))
````

## Concept 12: Detecting and Removing Outliers

### Method 1 — IQR (Interquartile Range):
```python
Q1 = df["Column"].quantile(0.25)
Q3 = df["Column"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Detect outliers
outliers = df[
    (df["Column"] < lower_bound) | 
    (df["Column"] > upper_bound)
]

# Remove outliers
df_clean = df[
    (df["Column"] >= lower_bound) & 
    (df["Column"] <= upper_bound)
]
```

### Method 2 — Z-Score:
```python
from scipy import stats

z_scores = stats.zscore(df["Column"])
outliers = df[abs(z_scores) > 3]
df_clean = df[abs(z_scores) <= 3]
```

### IQR vs Z-Score:
| Situation                          | Use     |
|------------------------------------|---------|
| Skewed or unknown distribution     | IQR     |
| Roughly normal distribution        | Z-Score |
| Simple and robust — default choice | IQR     |

### Key Points:
- IQR = Q3 - Q1 (middle 50% spread of data)
- 1.5 * IQR is the standard Tukey convention
- Always investigate outliers before removing them
- Outliers are sometimes the most important data points
- After removing, always compare shape before and after

````python
#Task-12:
"""
Using stock_data and the No.of Shares column:
Calculate Q1, Q3, IQR, lower bound, and upper bound
Print how many outliers exist
Print the dates on which outliers occurred
Remove the outliers and confirm the new shape
"""

Q1 = stock_data["No.of Shares"].quantile(0.25)
Q3 = stock_data["No.of Shares"].quantile(0.75)
IQR = Q3-Q1

lower_bound = Q1-1.5*IQR
upper_bound = Q3+1.5*IQR

outliers = stock_data[(stock_data["No.of Shares"] <lower_bound) | (stock_data["No.of Shares"]> upper_bound)]

print("No. of outliers: ",outliers.shape[0])
print(outliers["Date"])

print("Original Shape: ",stock_data.shape)

stock_data = stock_data[(stock_data["No.of Shares"]>=lower_bound) & (stock_data["No.of Shares"]<= upper_bound)]
print("New Shape: ",stock_data.shape)
````


## Extra Concept: apply()

### Syntax:
```python
df.apply(function, axis=0)   # Apply to each column
df.apply(function, axis=1)   # Apply to each row
```

### With lambda:
```python
df["Total"] = df.apply(lambda row: row["Col1"] + row["Col2"], axis=1)
```

### With named function:
```python
def classify(row):
    if row["Score"] >= 80:
        return "High"
    else:
        return "Low"

df["Category"] = df.apply(classify, axis=1)
```

### Key Points:
- axis=1 means function receives one row at a time
- axis=0 means function receives one column at a time
- axis=1 is used far more often in practice
- Always assign result to a new column

````python
#Task-E1:
df = pd.DataFrame({
    "Student": ["A", "B", "C", "D"],
    "Math":    [70, 45, 90, 60],
    "Science": [75, 40, 85, 55]
})
""" 
Create a column Average using apply() row-wise
Create a column Grade where Average >= 80 is Distinction, >= 60 is First, otherwise Second
"""

df["Average"] = df.apply(lambda row: (row["Math"] + row["Science"])/2, axis=1)

def grade(row):
    avg = (row["Average"])
    if avg >= 80:
        return "Distinction"
    elif avg >= 60:
        return "First"
    else:
        return "Second"

df["Grade"] = df.apply(grade, axis=1)
print(df)
````

## Extra Concept: dropna(thresh=) and dropna(axis=1)

### thresh= parameter:
```python
df.dropna(thresh=n)   # Keep rows with AT LEAST n non-null values
```
- Rows with fewer than n valid values are dropped
- Useful when you want to keep partially filled rows

### axis=1 — drop columns instead of rows:
```python
df.dropna(axis=1)              # Drop columns with ANY null value
df.dropna(axis=1, thresh=n)    # Drop columns with fewer than n non-null values
```

### Combining both:
```python
df.dropna(thresh=3, axis=1)    # Drop columns with fewer than 3 non-null values
```

### Default behaviour reminder:
```python
df.dropna()              # axis=0 — drops rows with ANY null value
df.dropna(how="all")     # drops rows only if ALL values are null
df.dropna(how="any")     # drops rows if ANY value is null (default)
```

### Key Points:
- axis=0 operates on rows (default)
- axis=1 operates on columns
- thresh overrides how parameter
- Always check shape before and after to confirm effect

## Extra Concept: groupby().size() and groupby().nth()

### size() — count rows per group:
```python
df.groupby("Col").size()           # Count of rows per group
df.groupby("Col").size()["Group"]  # Count for a specific group
```

### Difference between size() and count():
- size()  → counts ALL rows including NaN
- count() → counts only NON-NULL values per column

### nth() — select nth row from each group:
```python
df.groupby("Col").nth(0)    # First row of each group
df.groupby("Col").nth(1)    # Second row of each group
df.groupby("Col").nth(-1)   # Last row of each group
```

### Common pattern — find max/min per group using sort + nth:
```python
# Highest value per group
df.sort_values("Value").groupby("Col").nth(-1).reset_index()

# Lowest value per group
df.sort_values("Value").groupby("Col").nth(0).reset_index()
```

### Key Points:
- nth() is zero-indexed (0 = first, -1 = last)
- After nth(), group column becomes index — use reset_index()
- size() returns a Series, not a DataFrame
- Prefer sort + nth(-1) over idxmax() for finding group maximums

## Extra Concept: merge() with indicator=True

### Syntax:
```python
result = pd.merge(df1, df2, on="ID", how="outer", indicator=True)
```

### _merge column values:
| Value        | Meaning                          |
|--------------|----------------------------------|
| both         | Row exists in both DataFrames    |
| left_only    | Row exists only in left df       |
| right_only   | Row exists only in right df      |

### Counting by merge status:
```python

(result["_merge"] == "left_only").sum()
(result["_merge"] == "right_only").sum()
```

### Key Points:
- indicator=True only works with merge(), not concat()

- _merge column is of type Categorical, not plain string
- Use with how="outer" to see all unmatched rows clearly

````python
#Task-E5:
students = pd.DataFrame({
    "ID":   [1, 2, 3, 4],
    "Name": ["A", "B", "C", "D"]
})

marks = pd.DataFrame({
    "ID":    [2, 3, 5],
    "Score": [85, 90, 75]
})

""" 
Perform outer merge with indicator=True
Count how many rows are left_only, right_only, and both
Concatenate the two DataFrames row-wise using outer join and count total missing values
"""

result = pd.merge(students,marks,on="ID",how="outer",indicator=True)
print(result)

print((result["_merge"] == "both").sum())
print((result["_merge"] == "left_only").sum())
print((result["_merge"] == "right_only").sum())

result = pd.concat([students, marks], axis=0, ignore_index=True)
print(result.isnull().sum().sum())
````
