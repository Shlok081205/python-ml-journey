## Concept 4: Cleaning Data

### 4.1 Checking and Handling Missing Values:
```python
df.isnull()                          # True/False for every cell
df.isnull().sum()                    # Missing value count per column
df.dropna()                          # Drop rows with any missing value
df.dropna(subset=["ColumnName"])     # Drop rows where specific column is missing
df.fillna(0)                         # Fill missing with 0
df.fillna(df.mean(numeric_only=True)) # Fill missing with column mean
```

### 4.2 Handling Duplicates:
```python
df.duplicated().sum()   # Count duplicate rows
df.drop_duplicates()    # Remove duplicate rows
```

### 4.3 Dropping Columns or Rows:
```python
df = df.drop(columns=["ColumnName"])  # Drop a column
df = df.drop(index=2)                 # Drop a row by index number
```

### Key Notes:
- Always check isnull().sum() before any analysis
- inplace=True modifies original — reassigning (df = df.drop(...))
  is safer and more readable
- Date columns loaded as object type need pd.to_datetime() conversion

## Concept 5: Selecting Data with loc[] and iloc[]

### loc[] — Select by Label:
```python
df.loc[0]                        # Single row
df.loc[0:3]                      # Range of rows
df.loc[0:3, ["Col1", "Col2"]]    # Rows and specific columns
df.loc[df["Column"] > value]     # Rows matching a condition
df.loc[
    (df["Col1"] > val1) & 
    (df["Col2"] < val2)
]                                # Multiple conditions
```

### iloc[] — Select by Position:
```python
df.iloc[0]           # First row
df.iloc[0:5]         # First 5 rows
df.iloc[0:5, 0:3]    # First 5 rows, first 3 columns
```

### Key Difference:
- loc[]  → uses labels (column names, index labels)
- iloc[] → uses integer positions (0, 1, 2...)

### Professional Habit:
Always use loc[] over plain indexing (df[0:3]).
It is explicit, readable, and unambiguous.

````python
print('Single Row:',stock_data.loc[0])

print('Multiple Rows:',stock_data.loc[0:3])

print("Specific Rows and Columns:",stock_data.loc[0:3, ["Date", "Close Price"]])

print("Rows Satisfying the Condition:",stock_data.loc[stock_data["Close Price"] > 2700])

print('Rows Satisfying Multiple Condition:',stock_data.loc[(stock_data["Close Price"] > 2700) & (stock_data["No.of Shares"] > 200000)])
````


## Concept 8: groupby() and Aggregation

### Basic syntax:
```python
df.groupby("column")["target"].aggregation()
```

### Common aggregations:
```python
df.groupby("Col")["Target"].mean()
df.groupby("Col")["Target"].sum()
df.groupby("Col")["Target"].max()
df.groupby("Col")["Target"].min()
df.groupby("Col")["Target"].count()
```

### Multiple aggregations at once:
```python
df.groupby("Col")["Target"].agg(["mean", "min", "max", "std"])

# With custom column names
df.groupby("Col")["Target"].agg(
    Average="mean",
    Highest="max",
    Lowest="min"
)
```

### Group by multiple columns:
```python
df.groupby(["Col1", "Col2"])["Target"].mean()
```

### Always reset index for clean output:
```python
result = df.groupby("Col")["Target"].mean().reset_index()
```

### Working with dates — always convert first:
```python
df["Date"] = pd.to_datetime(df["Date"])
df["Month"] = df["Date"].dt.month
df["Year"] = df["Date"].dt.year
df["Month_Name"] = df["Date"].dt.strftime("%B")
```

### Key Points:
- groupby() splits data into groups, applies a function, combines results
- agg() lets you apply multiple functions and rename columns
- Date column must be datetime type before extracting month/year
- reset_index() converts group labels back into regular columns
- idxmax() on a grouped result finds the group with the highest value

```
### Your second question — where do min and max come from?

You are not finding a column called min or max. You are **applying the min and max functions** to the existing `Close Price` values within each group.

Think of it this way. After `groupby("Year")`, Pandas internally creates two buckets:
```
Year 2025 bucket:
[3624.95, 3568.00, 3529.95, 3483.95 ... all 2025 closing prices]

Year 2026 bucket:
[2636.40, 2647.70, 2629.50, 2574.50 ... all 2026 closing prices]
```

Then `agg()` runs your chosen functions on each bucket:
```
2025 bucket → mean() → 3256.37
2025 bucket → max()  → 3657.70
2025 bucket → min()  → 2889.00

2026 bucket → mean() → 3008.24
2026 bucket → max()  → 3294.45
2026 bucket → min()  → 2574.50


## Concept 9: merge() and concat()

### concat() — Stack DataFrames:
```python
# Vertical (more rows)
pd.concat([df1, df2], axis=0, ignore_index=True)

# Horizontal (more columns)
pd.concat([df1, df2], axis=1)
```
- Always use ignore_index=True when stacking vertically
  to avoid duplicate index values

### merge() — Join on a Common Column:
```python
pd.merge(df1, df2, on="common_column", how="join_type")
```

### Join Types:
| how    | Result                                          |
|--------|-------------------------------------------------|
| inner  | Only rows present in BOTH DataFrames            |
| left   | All rows from left, NaN where right has no match|
| right  | All rows from right, NaN where left has no match|
| outer  | All rows from both, NaN where either has no match|

### Key Points:
- merge() is equivalent to SQL JOIN
- inner join is the safest — no NaN values introduced
- Always run isnull().sum() after a merge to check for NaN
- outer join can introduce many NaN values unexpectedly
- on= parameter must be a column that exists in both DataFrames

````python
#Task-9:
"""
Stack both DataFrames vertically using concat() with ignore_index=True
Merge both DataFrames using an inner join on Product
Merge both DataFrames using a left join on Product and check for NaN values
"""

q1_sales = pd.DataFrame({
    "Product": ["Laptop", "Phone", "Tablet"],
    "Q1_Revenue": [150000, 80000, 45000]
})

q2_sales = pd.DataFrame({
    "Product": ["Phone", "Tablet", "Watch"],
    "Q2_Revenue": [95000, 50000, 30000]
})


print(pd.concat([q1_sales,q2_sales],axis=0,ignore_index=True),"\n")

print(pd.merge(q1_sales,q2_sales,on="Product",how="inner"),"\n")

print(pd.merge(q1_sales,q2_sales,on="Product",how="left").isnull().sum())
````

## Concept 10: Two-Way Cross Tabulation

### Basic syntax:
```python
pd.crosstab(df["Row_Variable"], df["Column_Variable"])
```

### With totals:
```python
pd.crosstab(df["Row"], df["Col"], margins=True)
```

### Normalized (proportions):
```python
# Whole table proportions
pd.crosstab(df["Row"], df["Col"], normalize=True)

# Row-wise proportions (each row sums to 1)
pd.crosstab(df["Row"], df["Col"], normalize="index")

# Column-wise proportions (each column sums to 1)
pd.crosstab(df["Row"], df["Col"], normalize="columns")
```

### Key Points:
- crosstab is used for two CATEGORICAL variables
- margins=True adds row and column totals
- normalize="index" shows pass/fail rate per category — 
  useful for comparing groups of different sizes
- Foundation of chi-square statistical tests

````python
#Task-10:
data = pd.DataFrame({
    "Gender":     ["Male","Female","Male","Female","Male",
                   "Female","Male","Female","Male","Female"],
    "Department": ["CE","IT","CE","CE","IT",
                   "IT","CE","IT","CE","CE"],
    "Result":     ["Pass","Pass","Fail","Pass","Pass",
                   "Fail","Pass","Pass","Fail","Pass"]
})
"""
Create a cross tabulation between Gender and Result
Add margins to see totals
Create a cross tabulation between Department and Result normalized by row — so you can see pass/fail rate per department as a proportion
"""

print(pd.crosstab(data["Gender"],data["Result"],margins=True),"\n")

print(pd.crosstab(data["Department"],data["Result"],normalize="index"))
````

## Concept 11: Correlation with corr()

### Correlation matrix for all numeric columns:
```python

```

### Correlation between two specific columns:

### Find column most correlated with a target:
```python
corr_matrix = df.corr(numeric_only=True)
```

### Correlation scale:
| Value       | Meaning                        |
|-------------|--------------------------------|
|  1.0        | Perfect positive correlation   |
|  0.5 to 1.0 | Strong positive correlation    |
|  0.0        | No linear relationship         |
| -0.5 to -1.0| Strong negative correlation    |
| -1.0        | Perfect negative correlation   |

### Key Points:
- Always use abs() when finding strongest correlation
  because -0.9 is just as strong as +0.9
- Diagonal of correlation matrix is always 1.0
- corr() only measures LINEAR relationships
- Correlation does not imply causation
- numeric_only=True avoids errors from text columns

## Concept 12: Detecting and Removing Outliers

### Method 1 — IQR (Interquartile Range):
```python
Q1 = df["Column"].quantile(0.25)
Q3 = df["Column"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Detect outliers
outliers = df[
    (df["Column"] < lower_bound) | 
    (df["Column"] > upper_bound)
]

# Remove outliers
df_clean = df[
    (df["Column"] >= lower_bound) & 
    (df["Column"] <= upper_bound)
]
```

### Method 2 — Z-Score:
```python
from scipy import stats

z_scores = stats.zscore(df["Column"])
outliers = df[abs(z_scores) > 3]
df_clean = df[abs(z_scores) <= 3]
```

### IQR vs Z-Score:
| Situation                          | Use     |
|------------------------------------|---------|
| Skewed or unknown distribution     | IQR     |
| Roughly normal distribution        | Z-Score |
| Simple and robust — default choice | IQR     |

### Key Points:
- IQR = Q3 - Q1 (middle 50% spread of data)
- 1.5 * IQR is the standard Tukey convention
- Always investigate outliers before removing them
- Outliers are sometimes the most important data points
- After removing, always compare shape before and after
